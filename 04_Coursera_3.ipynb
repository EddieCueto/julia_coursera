```julia
using Plots
```

```julia
r = 5
x0, y0 = 1,2
```

```julia
function trialcircle(r,x0,y0)
    t = range(0,stop=2Ï€,length=200)
    #x,y = map(cos,t), map(sin,t)
    x,y = r*cos.(t), r*sin.(t)
    x,y = x0 .+ x, y0 .+ y
    plot(x,y)
end
```

```julia
trialcircle(r,x0,y0)
```

```julia
function trialcircle!(r,x0,y0)
    t = range(0,stop=2Ï€,length=200)
    #x,y = map(cos,t), map(sin,t)
    x,y = r*cos.(t), r*sin.(t)
    x,y = x0 .+ x, y0 .+ y
    plot!(x,y)
end
```

```julia
trialcircle(1,0,0)
trialcircle!(2,0,0)
```

```julia
trialcircle(0,0,0)

for r = .1:.1:4
    trialcircle!(r,0,0)
end
plot!()
```

```julia
plot!(grid=false,legend=false,aspect_ratio=1)
```

```julia
trialcircle(0,0,0)

for r = .1:.1:4
    trialcircle!(r,r,0)
end
plot!(grid=false,legend=false,aspect_ratio=1)
```
